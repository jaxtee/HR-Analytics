## Visualize the Categorical Columns in relation to attrition rate

```python
# Plot BusinessTravel vs Attrition
sns.countplot(x="BusinessTravel", data=hr, hue="Attrition")
plt.title("Business Travel vs Attrition")
```

```python
hr.groupby(["Attrition", "BusinessTravel"])["BusinessTravel"].count()
```

```python
Travel_Rarely_Yes = (156 / len(hr[hr["BusinessTravel"] == "Travel_Rarely"])) * 100
Travel_Rarely_No = (887 / len(hr[hr["BusinessTravel"] == "Travel_Rarely"])) * 100
Travel_Frequently_Yes = (69 / len(hr[hr["BusinessTravel"] == "Travel_Frequently"])) * 100
Travel_Frequently_No = (208 / len(hr[hr["BusinessTravel"] == "Travel_Frequently"])) * 100
Non_Travel_Yes = (12 / len(hr[hr["BusinessTravel"] == "Non-Travel"])) * 100
Non_Travel_No = (138 / len(hr[hr["BusinessTravel"] == "Non-Travel"])) * 100

print(f"Non Travellers:\nAtrrition rate: {Non_Travel_Yes:.0f}%\n------\nRare Travellers:\nAtrrition rate: {Travel_Rarely_Yes:.0f}%\n------\nFrequent Travellers:\nAtrrition rate: {Travel_Frequently_Yes:.0f}%")
```

### Observation:
#### * While the plot shows that we have the highest attrition from staff who travel rarely and lowest from staff who don't travel, it is shown by the attrition rate for staff who dont travel is 8%, staff who rarely travel is 15% and frequent travellers 25%.
#### * Non travellers are most likely not to attrite.
#### * Frequent Flyers are most likely to attrite.
